### 49. Write a Python script to display the
- a. Current date and time
- b. Current year
- c. Month of year
- d. Week number of the year
- e. Weekday of the week
- f. Day of year
- g. Day of the month
- h. Day of week

```python
import datetime as dt
#a
current_datetime = dt.datetime.now()
print('Current_date_and_time:',current_datetime)

#b
current_year = dt.datetime.now().year
print('Current_Year:',current_year)

#c
current_month = dt.datetime.now().strftime("%B")
print('Month_of_the_year:',current_month)

#d
current_week_number = dt.datetime.now().strftime("%U")
print('Week_number_of_the_year:',current_week_number)

#e
current_weekday = dt.datetime.now().weekday()
weekday_names = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
current_weekday_name = weekday_names[current_weekday]
print('Weekday_of_the_week:',current_weekday_name)

#f
current_day_of_the_year = dt.datetime.now().strftime("%j")
print('Day_of_the_year:',current_day_of_the_year)

#g
current_day_of_the_month = dt.datetime.now().strftime("%d")
print('Day_of_the_month:',current_day_of_the_month)

#h
current_day_of_the_week = dt.datetime.now().strftime("%A")
print('Day_of_the_week:',current_day_of_the_week)
```

### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
from datetime import datetime
sample_string = 'Jul 1 2014 2:43PM'
datetime_object = datetime.strptime(sample_string, '%b %d %Y %I:%M%p')
formatted_datetime = datetime_object.strftime('%Y-%m-%d %H:%M:%S')
print("Expected Output:", formatted_datetime) 
```
